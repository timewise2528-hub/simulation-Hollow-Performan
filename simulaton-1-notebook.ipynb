````python
# Environment check
import sys, platform, subprocess
print('Python:', sys.version.split()[0])
print('Platform:', platform.system(), platform.release())
try:
    import pip
    print('pip:', pip.__version__)
except Exception:
    pass
````

````python
# Programmatic creation example (nbformat)
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell, new_code_cell

nb = new_notebook()
nb.cells = [
    new_markdown_cell('# Small generated notebook'),
    new_code_cell('print("Hello from generated notebook")')
]
# nbformat.write(nb, 'generated.ipynb')
````

````python
# Simulation example: Monte Carlo estimate of pi
import numpy as np
import matplotlib.pyplot as plt

N = 20000
x = np.random.rand(N)
y = np.random.rand(N)
inside = (x**2 + y**2) <= 1.0
pi_est = 4 * inside.sum() / N
print('Monte Carlo pi estimate:', pi_est)

plt.figure(figsize=(4,4))
plt.scatter(x[inside], y[inside], s=1, color='tab:blue')
plt.scatter(x[~inside], y[~inside], s=1, color='tab:orange')
plt.title(f'Monte Carlo pi ≈ {pi_est:.5f}')
plt.show()
````

````python
# Execute programmatically (nbclient)
from nbclient import NotebookClient
import nbformat

nb = nbformat.read('simulaton-1-notebook.ipynb', as_version=4)
client = NotebookClient(nb, timeout=120)
client.execute()
# After execution, save outputs:
nbformat.write(nb, 'simulaton-1-notebook.ipynb')
print('Executed and saved simulaton-1-notebook.ipynb')
````
